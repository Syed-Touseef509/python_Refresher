```python
raw_products = [
    ("Laptop", "Electronics", "1000"),
    ("Mouse", "Electronics", 500),
    ("Keyboard", "Electronics", "700"),
    ("Chair", "Furniture", None),
    ("Table", "Furniture", "1500"),
    ("Laptop", "Electronics", "1000")  
]
clean_products = []

for product in raw_products:
    name = product[0]
    category = product[1]
    price = product[2]
    
    if price is None:
        continue
    price = float(price)

    clean_products.append((name, category, price))
    
unique_products = set()
for product in clean_products:
    unique_products.add(product)
    

grouped = {}

for product in unique_products:

    name = product[0]
    category = product[1]
    price = product[2]

    if category not in grouped:
        grouped[category] = []

    grouped[category].append((name, price))

print("this is the product summary:\n")

for category in grouped:
    print(category + ":")

    for item in grouped[category]:
        print(" ",item[0]," ",item[1])
    print()      

```
