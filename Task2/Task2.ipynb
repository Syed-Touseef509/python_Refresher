Student Result Processor: The academic department has CSV files of student records
in a folder. Some files are corrupted, some are missing, and some have invalid data.
Build a system that reads all files safely, flags failed students, and writes a clean results
file without ever crashing


```python
import os
import csv
result = []
files=["student1.csv","student2.csv","student3.csv"]
folder_path = "files"
for filename in files:
    try:
        file_path = os.path.join(folder_path,filename)
        with open(file_path,"r") as file:
            reader=csv.reader(file)

            for row in reader:
                try:
                    name=row[0]
                    marks=int(row[1])

                    if marks>=50:
                        status = "pass"
                    else:
                        status="fail"
                    result.append([name,marks,status])
                except ValueError:
                    
                    print("invalid marks in the file",filename)
    except FileNotFoundError:
        
        print("file not found error",filename)
with open("clean_results.csv","w",newline="") as file:
    writer=csv.writer(file)
    for res in result:
        writer.writerow(res)

print("Clean file created successfully")
            
    

```
