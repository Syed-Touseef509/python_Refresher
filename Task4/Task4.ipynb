E-Commerce Inventory Analyzer: A shop owner has a messy product list with
duplicates, missing prices, and mixed categories. Clean the data, group by category, find
the top 5 most expensive items per category, and produce a summary dictionary —
written in a Pythonic style using comprehensions throughout.


```python
products = [
    {"name": "Laptop", "price": 1200, "category": "Electronics"},
    {"name": "Mouse", "price": 25, "category": "electronics"},
    {"name": "Keyboard", "price": 45, "category": "Electronics"},
    {"name": "Monitor", "price": 300, "category": "Electronics"},
    {"name": "Headphones", "price": None, "category": "Electronics"},
    {"name": "Sofa", "price": 700, "category": "Furniture"},
    {"name": "Table", "price": 250, "category": "furniture"},
    {"name": "Chair", "price": 150, "category": "Furniture"},
    {"name": "Laptop", "price": 1200, "category": "Electronics"},  
]

cleaned_product=[
    {
        "name":p["name"],
        "price":p["price"],
        "category":p["category"].lower()       
        
    }
    for p in products if p["price"] is not None
]

Unique_products={
    (p["name"],p["category"]):p for p in cleaned_product
}.values()

grouped = {
    category: [
        p for p in Unique_products
        if p["category"] == category
    ]
    for category in {p["category"] for p in Unique_products}
}

summary = {
    category: sorted(items, key=lambda x: x["price"], reverse=True)[:5]
    for category, items in grouped.items()
}

print(summary)
```

HR Employee Ranker: HR wants employees ranked by performance score, then by
experience as a tiebreaker, then by name alphabetically. The system must also support
quickly searching for a specific employee by name from the sorted list.


```python
employees = [
    {"name": "Ali", "score": 90, "experience": 5},
    {"name": "Sara", "score": 95, "experience": 3},
    {"name": "Ahmed", "score": 90, "experience": 7},
    {"name": "Zara", "score": 95, "experience": 5},
    {"name": "Bilal", "score": 85, "experience": 6}
]

sort_employee=sorted(
    employees,
    key=lambda e: (-e["score"],-e["experience"],e["name"])
    
)

rank_employees=[
    {**emp,"rank":rank+1}
    for rank,emp in enumerate(sort_employee)
]

print("Ranked employees:\n")
for emp in rank_employees:
    print(emp)
    
employees_by_name=sorted(rank_employees,key= lambda e: e["name"])

def search_employee(name, emp_list):
    left = 0
    right = len(emp_list) - 1

    while left <= right:
        mid = (left + right) // 2

        if emp_list[mid]["name"] == name:
            return emp_list[mid]

        elif emp_list[mid]["name"] < name:
            left = mid + 1

        else:
            right = mid - 1

    return "Employee not found"

print("\nSearch Result:")
print(search_employee("Sara",employees_by_name))
```
