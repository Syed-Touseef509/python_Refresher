```python
from functools import reduce
#Raw orders
orders = [
    {"id": 1, "category": "electronics", "amount": 1000},
    {"id": 2, "category": "clothing", "amount": 500},
    {"id": 3, "category": "electronics", "amount": 200},
    {"id": 4, "category": "grocery", "amount": 300}
]

discount_rules={
    "electronics":0.10,
    "clothing":0.20,
    "grocery":0.05
}

Minimum_Threshold=250

processed_orders=list(map(
    lambda order: {
            "id": order["id"],
            "category": order["category"],
            "original_amount": order["amount"],
            "final_amount": order["amount"] * (1 - discount_rules.get(order["category"], 0))
        },
        orders
))

filtered_orders=list(
    filter(
        lambda order: order["final_amount"]>=Minimum_Threshold,
        processed_orders 
    )
)

sorted_orders=sorted(filtered_orders,key=lambda order:order["final_amount"],reverse=True)

total_revenue=reduce(
    lambda total,order: total+order["final_amount"],
    sorted_orders,0
)

print("Filtred orders\n")
print(filtered_orders,"\n")
print("Total Revenue\n")
print("  ",total_revenue)
```
