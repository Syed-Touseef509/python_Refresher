```python
def stream_logs(file_path):
    with open(file_path,"r") as file:
        for line in file:
            yield line.strip()
            
def critical_error(log_stream):
    for line in log_stream:
        if "ERROR" in line or "CRITICAL" in line:
            yield line

def count_error(error_stream):
    count={}
    
    for line in error_stream:
        error_type=line.split(":")[0]
        count[error_type]=count.get(error_type,0)+1
        yield error_type,count[error_type]
log_stream=stream_logs("server.log")
error_stream=critical_error(log_stream)
counter_stream=count_error(error_stream)

for error ,count in counter_stream:
    print(f"{error} count ={count}")
```
