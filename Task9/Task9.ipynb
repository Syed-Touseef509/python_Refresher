```python

class Book:

    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True

    def __str__(self):
        if self.is_available:
            status = "Available"
        else:
            status = "Borrowed"

        return f"Book: {self.title} | Author: {self.author} | Status: {status}"



class Member:

    def __init__(self, name):
        self.name = name
        self.borrowed = []  

    def borrow_book(self, book):

        if not book.is_available:
            print("Sorry, this book is already borrowed.")
            return

        if len(self.borrowed) >= self.borrow_limit:
            print(f"{self.name} cannot borrow more books (limit reached).")
            return

        book.is_available = False
        self.borrowed.append(book)

        print(f"{self.name} borrowed '{book.title}'")

    def return_book(self, book, late_days=0):

        if book not in self.borrowed:
            print("This book was not borrowed by this member.")
            return

        self.borrowed.remove(book)
        book.is_available = True

        fine = self.calculate_fine(late_days)

        print(f"{self.name} returned '{book.title}'. Late fine = {fine}")

    def calculate_fine(self, late_days):
        return late_days * 1

    
    def __len__(self):
        return len(self.borrowed)

    def __str__(self):
        return f"Member: {self.name} | Books borrowed: {len(self)}"



class StandardMember(Member):

    def __init__(self, name):
        super().__init__(name)
        self.borrow_limit = 2

    
    def calculate_fine(self, late_days):
        return late_days * 2



class PremiumMember(StandardMember):

    def __init__(self, name):
        super().__init__(name)
        self.borrow_limit = 5

   
    def calculate_fine(self, late_days):
        return late_days * 1




book1 = Book("Python Programming1", "Guido van Rossum")
book2 = Book("Machine Learning Basics", "Andrew Ng")
book3 = Book("Deep Learning basic", "Jake VanderPlas")

user1 = StandardMember("Touseef")
user2 = PremiumMember("Haider")

print("\n--- Borrowing Books ---")
user1.borrow_book(book1)
user1.borrow_book(book2)
user1.borrow_book(book3)   
user2.borrow_book(book3)

print("\n--- Member Info ---")
print(user1)
print(user2)

print("\n--- Returning Books ---")
user1.return_book(book1, late_days=3)
user2.return_book(book3, late_days=3)
```
